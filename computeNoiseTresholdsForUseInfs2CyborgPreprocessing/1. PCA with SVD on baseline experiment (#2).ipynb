```scala
// ---- Specify path to csv file.
// Specify full system path
// or
// path starting from the directory of 
// notebook .ipynb file.

// the csv file is a direct ASCII export form
// Multi Channel DataManager for a given MEA.
// In MultiChannel DataManager, specify ASCII export
// to prepare the file for this notebook.
val dataPath = "data/2_MEA2_raw.csv"
val experimentName = "#2"
```

```scala
// ---- Get header as a string Array.
val header = spark
    .read.textFile(dataPath)
    .filter(_.contains("TimeStamp"))
    .flatMap(_.split(","))
    .collect
```

```scala
// ---- Load the data into a DataFrame
// with DataType as Int .
// DataFrame is the same as a Dataset with 
// import org.apache.spark.sql.Row
// as the rows.
import org.apache.spark.sql._
import org.apache.spark.sql.types._

// Pre-define how the data layout looks like.
// This leads to skipping of the first lines
// in the raw csv that do not match to the schema,
// if the read command is combined with the option
// DROPMALFORMED.
val schema = StructType(for (e <- header) yield {StructField(e, IntegerType, true)})

// load the data.
lazy val data = spark.read
    .schema(schema)
    .option("mode", "DROPMALFORMED")
    .csv(dataPath)
```

```scala
// Import libraries for data formats and SingularValueDecomposition
import org.apache.spark.ml.feature.VectorAssembler
import org.apache.spark.mllib.feature.StandardScaler
import org.apache.spark.sql.Row
import org.apache.spark.mllib.linalg.Vector
import org.apache.spark.mllib.linalg.Matrix
import org.apache.spark.mllib.linalg.distributed.RowMatrix
import org.apache.spark.mllib.linalg.SingularValueDecomposition

// set number of Principal Components
val numberPCs = 4

// ---- The following section prepares the data for SVD
// into a row-oriented distributed Matrix (RowMatrix)
// RowMatrix is the matrix form that the SVD implementation
// SingularValueDecomposition can handle.
// More info at: https://spark.apache.org/docs/2.2.0/mllib-dimensionality-reduction.html

// Required for the SVD implementation: Specify feature columns.
// they are all columns except TimeStamp and Reference electrode
val featureColumns = for(x <- header
                         if (x != "TimeStamp [µs]") && (x != "Ref (ID=14) [pV]")) yield {x}

// Required for the SVD implementation: Create an assembler that can transform the data
// row vectors of features
val assembler = new VectorAssembler()
    .setInputCols(featureColumns)
    .setOutputCol("Features")

// Required for the SVD implementation: transform the data
// row vectors of features
val dataAssembled = assembler.transform(data).select("Features")

// Required for the SVD implementation: 
// transform the data into Resilient Distributed Dataset (RDD)
// where each row is a org.apache.spark.mllib.linalg.Vector
val SVDDataVectors = dataAssembled.rdd.map{x:Row => 
    val denseVector = x.getAs[org.apache.spark.ml.linalg.SparseVector]("Features").toDense
    org.apache.spark.mllib.linalg.Vectors.fromML(denseVector)
    // fix: https://stackoverflow.com/questions/40109807/classcastexception-org-apache-spark-ml-linalg-densevector-cannot-be-cast-to-org
    }

// Mean centering
val scaler = new StandardScaler(withMean = true, withStd = true).fit(SVDDataVectors)
val SVDDataVectorsScaled = SVDDataVectors.map(x => scaler.transform(x))

// Required for the SVD implementation: transform the RDD data
// into a RowMatrix that is filled into the SVD implementation 
val SVDData = new RowMatrix(SVDDataVectorsScaled)

// ---- Finished preparing the data RowMatrix

// --- Compute the top numberPCs singular values
// and corresponding singular vectors.
val svd: SingularValueDecomposition[RowMatrix, Matrix] = SVDData.computeSVD(numberPCs, computeU = true)

val U: RowMatrix = svd.U  // The U factor is a RowMatrix.
val s: Vector = svd.s     // The singular values are stored in a local dense vector.
val V: Matrix = svd.V     // The V factor is a local dense matrix.
// ---- Finished compute SVD
```

```scala
// Prepare Singular values
// for histogram plot with Vegas
var singularValues = s.toArray.zipWithIndex.map{
    case (e, i) =>
    val singularValueNumber = i + 1
    Map("Singular Value" -> s"S: $singularValueNumber", "Magnitude" -> e)
    }
// Prepare loadings
val loadings = V.colIter.toList
```

```scala
// Plotting singular values in Vegas. Histpgram plot.
Vegas("Singular Values from raw data of " + experimentName)
    .withData(singularValues)
    .mark(Bar)
    .encodeX("Singular Value", Nominal, sortField=Sort("Magnitude", AggOps.Mean, SortOrder.Desc))
    .encodeY("Magnitude", Quantitative)
    .show
```

```scala
// Prepare loadings for plotting
// with Vegas
var inputData = loadings.zipWithIndex.map{
    case (e, i) =>
    val singularValueNumber = i + 1
    val combinations = for (value <- e.toArray) yield (
        Map("Loading PC" -> singularValueNumber, 
            "Electrode" -> featureColumns(e.toArray.indexOf(value)), 
            "Magnitude" -> value)
    )
    combinations
    }
    .flatten
    .toArray
```

```scala
// Plot Loadings with Vegas
Vegas("Loadings from raw data of " + experimentName)
    .withData(inputData)
    .mark(Line)
    .encodeY("Electrode", Nominal)//, sortField=Sort("Loading PC :", AggOps.Mean, SortOrder.Desc))
    .encodeX("Magnitude", Quantitative)
    .encodeColumn("Loading PC", Quantitative, enableBin=false)
    //.encodeDetailFields(Field(field="Loading PC: ", dataType=Quantitative))
    .show
```

```scala
// plot first 25 seconds of PC1 line score
val secondsInto = 1
val samplingRate = 10000

// zero-indexed columns. Column(3) is for PC1, when numberPCs = 4
val score1 = U.rows.map(v => v.apply(3)).take(secondsInto * samplingRate)

// Prepare loadings for plotting
// with Vegas
var inputDataScore1 = score1.zipWithIndex.map{ case (e, i) =>
        Map("Seconds" -> i/10000.0, 
            "Magnitude" -> e )
        }
        .toArray
// Plot Loadings with Vegas
Vegas("Score1 from raw data of " + experimentName, width=1000, height=500)
    .withData(inputDataScore1)
    .mark(Line)
    .encodeY("Magnitude", Quantitative)//, sortField=Sort("Loading PC :", AggOps.Mean, SortOrder.Desc))
    .encodeX("Seconds", Quantitative)
    //.encodeColumn("Loading PC", Quantitative, enableBin=false)
    //.encodeDetailFields(Field(field="Loading PC: ", dataType=Quantitative))
    .show
```

```scala
val secondsInto = 1
val goalSamplingRate = 100

val score1resampled = U.rows
    .map(v => v.apply(3))
    .zipWithIndex
    .filter(_._2 % (samplingRate / goalSamplingRate) == 0)
    .map(_._1)
    .take(secondsInto * goalSamplingRate)

// Prepare loadings for plotting
// with Vegas
var inputDataScore1 = score1resampled.zipWithIndex.map{ case (e, i) =>
        Map("Seconds" -> i/goalSamplingRate.toDouble, 
            "Magnitude" -> e )
        }
        .toArray
// Plot Loadings with Vegas
Vegas("Score1 from raw data of " + experimentName, width=1000, height=500)
    .withData(inputDataScore1)
    .mark(Line)
    .encodeY("Magnitude", Quantitative)//, sortField=Sort("Loading PC :", AggOps.Mean, SortOrder.Desc))
    .encodeX("Seconds", Quantitative)
    //.encodeColumn("Loading PC", Quantitative, enableBin=false)
    //.encodeDetailFields(Field(field="Loading PC: ", dataType=Quantitative))
    .show
```
